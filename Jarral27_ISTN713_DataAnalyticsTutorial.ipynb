```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('SE Test One.csv')


df['Scaled_Percentage'] = df['Percentage'] / 10

plt.figure(figsize=(10, 6))
plt.hist(df['Scaled_Percentage'], bins=20, range=(0, 10), color='purple', alpha=0.7, edgecolor='black')

plt.xlabel('')
plt.ylabel('Participants')
plt.title('')
plt.legend(['Participants'])
plt.xticks([i * 0.5 for i in range(21)])

plt.grid(True)
plt.show()
```
